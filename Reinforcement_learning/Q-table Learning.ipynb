```python
import numpy as np
import gym
```

```python
#loading the enviroment 
env = gym.make('FrozenLake-v0')
```

```python
#Learning theQ-table non-neural network way
Q = np.zeros([env.observation_space.n, env.action_space.n])
#setting all the initial values of the table to zero
lr = 0.8 #learning rate
y = 0.95
num_episodes = 2000

r_list = []
for i in range(num_episodes):
    s = env.reset()
    rAll = 0
    d = False
    j = 0
    while j < 99:
        j += 1
        a = np.argmax(Q[s,:] + np.random.randn(1, env.action_space.n)*(1./(i+1)))
        s1,r,d,_ = env.step(a)
        Q[s,a] = Q[s,a] + lr*(r+y*np.max(Q[s1,:])-Q[s,a])
        rAll += r
        s = s1
        if d == True:
            break
    r_list.append(rAll)
```

```python
print ("Score over time:" + str(sum(r_list)/num_episodes))
```
