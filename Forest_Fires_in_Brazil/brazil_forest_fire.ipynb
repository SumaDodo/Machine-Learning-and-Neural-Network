```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv" ,encoding = "ISO-8859-1")
df.head()
```

```python
yearly_number = pd.pivot_table(df, values="number", index =["year"], aggfunc=np.sum)
```

```python
yearly_number
```

```python
monthly_number = pd.pivot_table(df, values = "number", index=["month"], aggfunc=np.sum)
```

```python
monthly_number
```

```python
state_number = pd.pivot_table(df, values ="number", index=["state"], aggfunc =np.sum)
```

```python
state_number
```

```python
#plt.figure(figsize=(15,6))
plt.grid(state_number)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.plot(monthly_number)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.plot(yearly_number)
plt.xticks(rotation='vertical')
plt.show()
```
